# Preprocessing

### **`1`**
I began by loading the Iris dataset from a CSV file named iris.csv into a DataFrame called iris using the `pd.read_csv()` function. This function reads the CSV file and converts it into a DataFrame, which is a tabular data structure provided by the Pandas library.

Later, the script prints the first few rows of the dataset using the `head()` method to verify that the data has been loaded correctly and to inspect its structure and contents. This step ensures that the dataset has been successfully imported and can be used for further analysis and visualization.

```python
# Importing necessary libraries
import numpy as np  # Importing NumPy library for numerical operations
import pandas as pd  # Importing Pandas library for data manipulation and analysis
import seaborn as sns  # Importing Seaborn library for statistical data visualization
import matplotlib.pyplot as plt  # Importing Matplotlib library for plotting graphs

# Load the data from the CSV file
iris = pd.read_csv('iris.csv')  # Reading the iris dataset from the CSV file into a DataFrame called 'iris'

# Displaying the first few rows of the dataset to verify the data has been loaded correctly
print(iris.head())  # Printing the first few rows of the dataset to inspect the structure and contents
```

### **`2`**
Dataset Information:

- variety: The target or response column, representing the species of the iris flower, which I aim to predict based on other feature columns.

- sepal.width: The sepal width for the iris flower in centimeters.
- petal.length: The petal length for the iris flower in centimeters.
- petal.width: The petal width for the iris flower in centimeters.

### **`3`**

# **1 `Histogram`**
## Visualization 1.1


Firstly. to examine the data's distribution, `Histograms` are utilized. They offer insights into whether the data follows a normal or skewed distribution pattern. When examining individual column distributions, providing only the specific column is sufficient.

```python
# Using countplot to visualize the number of each kind of iris species flower
sns.countplot(x=iris['variety'])
plt.show()
```

To examine the entire DataFrame, I decided to pass the DataFrame itself.
However, when dealing with a larger number of columns or when one specifically require the distribution for a particular column,
It is advisable to analyze the histogram of each individual column separately.

## Visualization 1.2

```python
# Visualizing the distribution of data (Petal Length) using Histogram
iris['petal.length'].hist(bins=20)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()
```

To examine the distribution of `petal length` and `petal width` across different species in the Iris Flower Dataset and enhance clarity through additional visual enhancements, the following approach can be adopted:

## Visualization 1.3

```python
# Visualizing the distribution of data for the entire dataset
iris.hist(bins=20, figsize=(10,10))
plt.suptitle('Histograms of Iris Dataset Features', fontsize=16)
plt.show()
```

## Visualization 1.4
**`Final Visualization 1`**

```python
# Visualizing the distribution of Iris dataset features for different species
iris_subset = iris.groupby('variety')

# Plotting histograms for Petal Length, Petal Width, Sepal Length, and Sepal Width
plt.figure(figsize=(12, 10))

plt.suptitle('Distribution of Iris Dataset Features across 3 varieties', fontsize=20)  # Add this line for the overall subtitle

plt.subplot(221)
iris_subset['petal.length'].plot.hist(alpha=0.4)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length for different varieties')
plt.legend(loc=(0.69, 0.75))
plt.grid()

plt.subplot(222)
iris_subset['petal.width'].plot.hist(alpha=0.4)
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width for different varieties')
plt.legend(loc=(0.69, 0.75))
plt.grid()

plt.subplot(223)
iris_subset['sepal.length'].plot.hist(alpha=0.4)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length for different varieties')
plt.legend(loc=(0.69, 0.75))
plt.grid()

plt.subplot(224)
iris_subset['sepal.width'].plot.hist(alpha=0.4)
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width for different varieties')
plt.legend(loc=(0.69, 0.75))
plt.grid()

plt.tight_layout()
plt.show()
```

# **2 `Scatter Plots`**
## Visualization 2.1

Next, one can explore the connections between features by employing scatter plots. We have the option to examine individual scatterplots for various variables or utilize Seaborn's Pairplot to assess the relationships of multiple variables concurrently. However, it's not recommended to employ Pairplot when dealing with a large number of features. In such instances, it's preferable to use a subset of features for the Pairplot visualization.

```python
# Exploring connections between features using scatter plots
sns.set_style('darkgrid')

plt.figure(figsize=(10, 10))
plt.subplot(211)
sns.scatterplot(data=iris, x='petal.length', y='petal.width', hue='variety')
plt.title('Relationship between Petal Length and Petal Width', fontsize=20)

plt.subplot(212)
sns.scatterplot(data=iris, x='sepal.length', y='sepal.width', hue='variety')
plt.title('Relationship between Petal Length and Sepal Length', fontsize=20)

plt.tight_layout()
plt.show()
```

Given the relatively small size of this dataset and its inclusion of only four features, employing a pairplot allows us to examine the relationship between two variables categorized by the Species column.

## Visualization 2.2
**`Final Visualization 2`**

```python
# Using Pairplot to analyze relationships between features
sns.pairplot(data=iris, hue='variety')
plt.suptitle('Pairplot of Iris Dataset Features by Variety', fontsize=10, ha='center', va='bottom')
plt.show()
```

# **3 `Box Plots`**
## Visualization 3.1

`Box plots` serve as an alternative method for visualizing distributions, offering a more comprehensive overview compared to `histograms`. They showcase various statistical measures such as the minimum value, maximum value, first and third quartiles, median, interquartile range (IQR), and identification of outliers. Additionally, box plots are beneficial when exploring distributions alongside categorical variables. 

```python
# Visualizing the distribution of Sepal Length for the Versicolor variety using box plots
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, y='sepal.length')
plt.title('Boxplot of Sepal Length for Versicolor Variety', fontsize=16)
plt.xlabel('Versicolor Variety', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)
plt.show()
```

## Visualization 3.2
**`Final Visualization 3`**

```python
# Plotting histograms for Petal Length, Petal Width, Sepal Length, and Sepal Width
plt.figure(figsize=(12, 10))

plt.suptitle('Comparing the distribution of Iris dataset features for different species using box plots', fontsize=20)  # Add this line for the overall subtitle


# Boxplot of Sepal Length by Species
plt.subplot(221)
sns.boxplot(data=iris, x='variety', y='sepal.length', hue='variety')
plt.title('Boxplot of Sepal Length by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)

# Boxplot of Sepal Width by Species
plt.subplot(222)
sns.boxplot(data=iris, x='variety', y='sepal.width', hue='variety')
plt.title('Boxplot of Sepal Width by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)

# Boxplot of Petal Length by Species
plt.subplot(223)
sns.boxplot(data=iris, x='variety', y='petal.length', hue='variety')
plt.title('Boxplot of Petal Length by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)

# Boxplot of Petal Width by Species
plt.subplot(224)
sns.boxplot(data=iris, x='variety', y='petal.width', hue='variety')
plt.title('Boxplot of Petal Width by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)

plt.tight_layout()
plt.show()
```

# **4 `Violin Plots`**
## Visualization 4.1

It's worth noting that Violin Plots share similarities with Box Plots in terms of visualizing distributions.

```python
# Visualizing the distribution of Sepal Length for the Versicolor variety using box plots
plt.figure(figsize=(8, 6))
sns.violinplot(data=iris, y='sepal.length')
plt.title('Boxplot of Sepal Length for Versicolor Variety', fontsize=16)
plt.xlabel('Versicolor Variety', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)
plt.show()
```

## Visualization 4.2
**`Final Visualization 4`**

```python
# Plotting histograms for Petal Length, Petal Width, Sepal Length, and Sepal Width
plt.figure(figsize=(12, 10))

plt.suptitle('Comparing the distribution of Iris dataset features for different species using violin plots', fontsize=20)  # Add this line for the overall subtitle


# Boxplot of Sepal Length by Species
plt.subplot(221)
sns.violinplot(data=iris, x='variety', y='sepal.length', hue='variety')
plt.title('Violinplot of Sepal Length by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)

# Boxplot of Sepal Width by Species
plt.subplot(222)
sns.violinplot(data=iris, x='variety', y='sepal.width', hue='variety')
plt.title('Violinplot of Sepal Width by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)

# Boxplot of Petal Length by Species
plt.subplot(223)
sns.violinplot(data=iris, x='variety', y='petal.length', hue='variety')
plt.title('Violinplot of Petal Length by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)

# Boxplot of Petal Width by Species
plt.subplot(224)
sns.violinplot(data=iris, x='variety', y='petal.width', hue='variety')
plt.title('Violinplot of Petal Width by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)

plt.tight_layout()
plt.show()
```
